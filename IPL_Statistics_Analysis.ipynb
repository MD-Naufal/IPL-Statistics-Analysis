# Top 10 batsmen based on past 3 years batting average. (min 60 balls faced in each season)

```python
# Convert 'balls' and 'runs' to numeric values in case they are strings
batsmen_data['balls'] = pd.to_numeric(batsmen_data['balls'], errors='coerce')
batsmen_data['runs'] = pd.to_numeric(batsmen_data['runs'], errors='coerce')
# Filter players who have faced at least 60 balls in each season
filtered_data = batsmen_data.groupby('batsmanName').filter(lambda x: x['balls'].sum() >= 60)
filtered_data
```

```python
# Calculate the total runs and total balls for each batsman, and then their average
batsman_stats = filtered_data.groupby('batsmanName').agg({
    'runs': 'sum',
    'balls': 'sum',
    'out/not_out': lambda x: (x == 'out').sum()  # Count the number of dismissals
}).reset_index()

batsman_stats
```

```python
# Calculate batting average = total runs / number of dismissals
# Note: If a player has no dismissals, we will avoid division by zero
batsman_stats['batting_average'] = batsman_stats.apply(
    lambda row: row['runs'] / row['out/not_out'] if row['out/not_out'] > 0 else row['runs'], axis=1
)

# Sort by batting average in descending order to get the top 10
top_10_batsmen_bat_avg = batsman_stats.sort_values(by='batting_average', ascending=False).head(10)

top_10_batsmen_bat_avg[['batsmanName', 'batting_average','runs','balls']]
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

plt.figure(figsize=(12, 7))
bars = sns.barplot(x=top_10_batsmen_bat_avg['batsmanName'], y=top_10_batsmen_bat_avg['batting_average'], palette='flare')
plt.xlabel('Batsmen')
plt.ylabel('Batting Average')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 batsmen based on past 3 years batting average. (min 60 balls faced in each season)')
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
```

```python
filtered_bowler_data = bowler_grouped[bowler_grouped['total_balls'] >= 60]
filtered_bowler_data
```

```python
plt.figure(figsize=(12, 7))
bowlers = top_10_bowlers_avg['bowlerName']
averages = top_10_bowlers_avg['bowling_average']


plt.bar(bowlers, averages, color='lightblue')

for i, avg in enumerate(averages):
    plt.text(i, avg - 0.5, f'{avg:.2f}', ha='center', va='top', color='black')


plt.ylabel('Bowling Average')
plt.title('Top 10 bowlers based on past 3 years bowling average (min 60 balls bowled)')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

# Top 5 batsmen based on past 3 years boundary % (fours and sixes)

```python
batsman_stats = filtered_data.groupby('batsmanName').agg({
    'runs': 'sum',
    'balls': 'sum',
    '4s': 'sum',
    '6s': 'sum',
    'out/not_out': lambda x: (x == 'out').sum()  # Count the number of dismissals
}).reset_index()

batsman_stats
```

```python
# Calculate boundary runs and boundary percentage
batsman_stats['boundary_runs'] = batsman_stats['4s'] * 4 + batsman_stats['6s'] * 6
batsman_stats['boundary_percentage'] = (batsman_stats['boundary_runs'] / batsman_stats['runs']) * 100

# Sort by boundary percentage in descending order and select the top 5
top_5_batsmen_boundary = batsman_stats.sort_values(by='boundary_percentage', ascending=False).head(5).reset_index()

top_5_batsmen_boundary
```

```python
plt.figure(figsize=(12, 7))
bowlers = top_5_batsmen_boundary['batsmanName']
averages = top_5_batsmen_boundary['boundary_percentage']


sns.barplot(x=top_5_batsmen_boundary['batsmanName'], y=top_5_batsmen_boundary['boundary_percentage'], palette='magma' )

for i, avg in enumerate(top_5_batsmen_boundary['boundary_percentage']):
    plt.text(i, avg + 0.5, f'{avg:.2f}', ha='center', va='bottom', color='black')


plt.ylabel('Boundary percentage')
plt.title('Top 5 batsmen based on past 3 years boundary % (fours and sixes).')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

# Top 4 teams based on past 3 years winning %

```python
team_data = r'C:\Users\moham\Documents\IPL Statistics Analysis Dataset\dim_match_summary.csv'
teams_data = pd.read_csv(team_data)
teams_data.head(5)
```

```python
# Convert matchDate to datetime
teams_data['matchDate'] = pd.to_datetime(teams_data['matchDate'], format='%d-%b-%y')

# Filter data for the past 3 years
three_years_ago = teams_data['matchDate'].max() - pd.DateOffset(years=3)
filtered_data = teams_data[teams_data['matchDate'] >= three_years_ago]

```

```python
# Calculate the number of matches played and won by each team
matches_played = filtered_data['team1'].value_counts() + filtered_data['team2'].value_counts()
matches_won = filtered_data['winner'].value_counts()

matches_played
matches_won
```

```python
win_percentage = (matches_won / matches_played) * 100

top_4_teams = win_percentage.nlargest(4)

print("Top 4 teams by winning percentage:")
top_4_teams
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
bars = sns.barplot(top_4_teams, palette='rocket_r')
plt.xlabel('TEAMS')
plt.ylabel('Win Percentage')
plt.xticks(rotation=45, ha='right')  
plt.title('Top 4 teams by winning percentage')
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.2f}', ha='center', va='bottom')
plt.tight_layout() 
plt.show()
```

# Top 2 teams with the highest number of wins achieved by chasing targets over the past 3 years

```python
filtered_data['won_by_chase'] = filtered_data['margin'].apply(lambda x: 'wickets' in str(x))

chasing_wins = filtered_data[filtered_data['won_by_chase']].groupby('winner').size()

chasing_wins
```

```python
top_2_chasing_teams = chasing_wins.nlargest(2)

print("Top 2 teams with the most wins by chasing:")
top_2_chasing_teams
```

```python
plt.figure(figsize=(10, 6))
bars = sns.barplot(x=top_2_chasing_teams.values, y=top_2_chasing_teams.index, palette='magma')

# Labeling the axes and chart title
plt.xlabel('Total Wins by Chasing')
plt.ylabel('Teams')
plt.title('Top 2 Teams with the Most Wins by Chasing')

# Add data labels to the bars, positioned slightly inside the bars
for bar in bars.patches:
    plt.text(bar.get_width() - 0.5, bar.get_y() + bar.get_height()/2, f'{bar.get_width():.0f}', ha='center', va='center', color='black')

# Adjust layout for better appearance
plt.tight_layout()
plt.show()
```
